```python
%matplotlib inline

import os

# let's not pollute this blog post with warnings
from warnings import filterwarnings
filterwarnings('ignore')

import numpy as np
import pandas as pd
import csv
import copy

#Import Library of Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
```

```python
labelpath = os.path.join('UCI_Credit_Card.csv')
train_labels = pd.read_csv(labelpath)
del train_labels['ID']
continuous_features = ['LIMIT_BAL', 'AGE', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']
train_labels = train_labels[continuous_features]
train_labels.to_csv('continuous_features.csv', index=False)
```

```python
discretized = copy.deepcopy(train_labels)
for col in train_labels:
    discretized[col] = pd.qcut(discretized[col], 10, labels=False, duplicates='drop')
discretized.to_csv('continuous_features_discretized.csv', index=False)
```

```python
discrete_features = pd.read_csv("discrete_features.csv")
result = pd.concat([discrete_features,discretized], axis=1)
result.to_csv('all_features_discretized.csv', index=False)
```
